```python
import matplotlib.pyplot as plt
```

```python
# show image
plt.imshow(img)
```

```python
recolor = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(recolor)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
plt.show()
```

# Write out Image

```python
cv2.imwrite("grayimage.jpg", gray)
```

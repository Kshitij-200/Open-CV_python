# Import Dependencies

```python
import cv2
import os
import matplotlib.pyplot as plt
```

# Load Image

```python
# Define a file path
IMG_PATH = os.path.join("Images","image.jpg")
```

```python
# Read in image
img = cv2.imread(IMG_PATH)
```

```python
img
```

```python
plt.imshow(img)
plt.show()
```

```python
# Resize Image
resized_img = cv2.resize(img, (int(img.shape[1]/2), int(img.shape[0]/2)))
# View image using open CV
cv2.imshow("Frame View", resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Edge Detection

```python
# Apply Canny
gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
blur = cv2.GaussianBlur(gray, (5,5), 0)
canny = cv2.Canny(blur, threshold1 = 205, threshold2 = 210)
```

```python
# Resize Image
resized_img = cv2.resize(canny, (int(img.shape[1]/2), int(img.shape[0]/2)))
# View image using open CV
cv2.imshow("Frame View", resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Export Edges

```python
cv2.imwrite("edges.jpg", canny)
```
